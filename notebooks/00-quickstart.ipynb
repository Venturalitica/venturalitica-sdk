# Venturalitica: 60-Second Quickstart

Welcome to **Venturalitica**! This notebook demonstrates the fastest way to run a bias audit on a real-world dataset.

### Goal:
Perform a fairness audit on the UCI German Credit dataset in just one line of code.

```python
import venturalitica as vl
print(f"Venturalitica version: {vl.__version__}")
```

## 1. Run the Quickstart Audit

The `vl.quickstart()` function automatically:
1. Fetches the German Credit dataset from UCI.
2. Loads a pre-defined Loan Fairness policy.
3. Executes the audit and displays a premium structured report.

```python
vl.quickstart('loan')
```

## What happened?
- **Data loaded**: 1000 samples of credit applicants.
- **Bias Detected**: The audit identified that the dataset has significant **Age Bias** (DI = 0.361), while **Gender Fairness** passed with a ratio of 0.836.
